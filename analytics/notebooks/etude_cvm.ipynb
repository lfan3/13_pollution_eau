# Etude sur le chlorure de vinyle monomère (CVM)


* Qu’est-ce que le chlorure de vinyle monomère (CVM) ?

C’est un gaz reconnu comme cancérigène certain pour l’humain par le Centre international de Recherche sur le Cancer et l’OMS, à partir de 1987. Il favoriserait deux formes de cancer du foie : l’angiosarcome hépatique, et le carcinome hépatocellulaire. Depuis les années 1930, les preuves de sa toxicité s’accumulaient.

* Pourquoi retrouve-t-on ce composant dans l’eau du robinet ?

À partir des années 1960, ce composant a été utilisé dans les canalisations d’eau en PVC, privilégiées pour raccorder des hameaux ou des fermes qui étaient alors coupées du réseau. Le PVC était alors privilégié à la place du plomb. Entre 140 000 km (selon le ministère de la Santé) et 340 000 km de canalisations en PVC (selon les délégataires de service public de l’eau) ont été installés à cette époque. L’utilisation de ce composant a été éliminé de la majeure partie des PVC dans les années 1980.

## Combien de communes ont un relevé de dépassement de seuil ?

* Pour cela on va merger les dataframes communes, prelevements et resultats. Il y a documentations en pdf pour voir les jointures sur ce lien https://www.data.gouv.fr/fr/datasets/r/36afc708-42dc-4a89-b039-7fde6bcc83d8

```python
pd.set_option("display.max_columns", None)  # Afficher toutes les colonnes
# pd.set_option('display.width', 1000)  # Ajuster la largeur de l'affichage
```

```python
# Jointure entre "communes" et "prelevements" sur "cdreseau"
merged_1 = pd.merge(communes, prelevements, on="cdreseau", how="left")

# Jointure entre merged_1 et "resultats" sur referenceprel
final_df = pd.merge(merged_1, resultats, on="referenceprel", how="left")
final_df.head()
```

```python
final_df.shape
```

```python
# On va filtrér les prélevements CVM
final_df_cvm = final_df[final_df["libminparametre"] == "Chlorure de vinyl monomère"]
final_df_cvm.shape
```

```python
# Filtrer les dépassements de seuil (> 0.5)
final_df_cvm_seuil = final_df_cvm[final_df_cvm["valtraduite"] > 0.5]

print(
    f"Nombre de prélevement avec un dépassement de seuil : {final_df_cvm_seuil.shape[0]}"
)
```

```python
# Identifier les communes qui dépassent le seuil (valtraduite > 0.5)
communes_depassement = final_df_cvm.groupby("inseecommune")["valtraduite"].max() > 0.5
# Nombre de communes dépassant le seuil
nombre_communes_depassement = communes_depassement.sum()

# Afficher le résultat
print(f"Nombre de communes qui dépassent le seuil : {nombre_communes_depassement}")
```

```python
import folium
import geopandas as gpd

# Charger les coordonnées des communes françaises (GeoJSON)
url_geojson = "https://raw.githubusercontent.com/gregoiredavid/france-geojson/master/communes.geojson"
gdf_communes = gpd.read_file(url_geojson)[["code", "nom", "geometry"]]
gdf_communes.rename(columns={"code": "inseecommune"}, inplace=True)
gdf_communes["inseecommune"] = gdf_communes["inseecommune"].apply(
    lambda x: x.lstrip("0")
)
gdf_communes["inseecommune"] = gdf_communes["inseecommune"].astype(str)
final_df_cvm["inseecommune"] = final_df_cvm["inseecommune"].astype(str)

# Identifier les communes qui dépassent le seuil (valtraduite > 0.5)
communes_depassement = final_df_cvm.groupby("inseecommune")["valtraduite"].max()

gdf_final = communes_depassement.reset_index().merge(
    gdf_communes, on="inseecommune", how="left"
)


# Filtrer uniquement les communes qui dépassent le seuil
gdf_final = gdf_final[gdf_final["valtraduite"] > 0.5]

# Créer une carte centrée sur la France
m = folium.Map(location=[46.603354, 1.888334], zoom_start=6)

# Ajouter uniquement les communes dépassant le seuil en rouge avec un popup détaillé
for _, row in gdf_final.iterrows():
    if row["geometry"] and not row["geometry"].is_empty:
        lon, lat = row["geometry"].centroid.x, row["geometry"].centroid.y
        popup_text = (
            f"<b>Commune :</b> {row['nom']}<br><b>Valeur :</b> {row['valtraduite']:.2f}"
        )
        folium.CircleMarker(
            location=[lat, lon],
            radius=5,
            color="red",
            fill=True,
            fill_color="red",
            fill_opacity=0.6,
            popup=folium.Popup(popup_text, max_width=250),
        ).add_to(m)

# Afficher la carte

m
```
